# Data Preprocessing

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/jabdurrahmaan/smart-meter-benchmark/main/datasets/daily_dataset_small.csv')
df
```

```python
# Move and rename column 'day', create column 'id'
df.insert(0, "id", range(1, len(df) + 1))

cols = df.columns.tolist()
cols.insert(0, cols.pop(cols.index('id')))
cols.insert(1, cols.pop(cols.index('day')))

df = df[cols]
df = df.rename(columns={'day': 'energy_date'})

df
```
