# Data Preprocessing

```python
import pandas as pd
```

```python
# Read CSV file
df = pd.read_csv('https://raw.githubusercontent.com/jabdurrahmaan/smart-meter-benchmark/main/datasets/daily_dataset_small.csv')
df

# Convert 'day' column to datetime format
df['day'] = pd.to_datetime(df['day'])

# Convert datetime format to desired format
df['day'] = df['day'].dt.strftime('%Y-%m-%d %H:%M:%S.%f%z')

# Move 'day' column to first position and rename
cols = df.columns.tolist()
cols.insert(0, cols.pop(cols.index('day')))
df = df[cols]
df = df.rename(columns={'day': 'time'})

# Print the modified DataFrame
df
```
