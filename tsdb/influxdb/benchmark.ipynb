# Benchmark of Compressed Dataset

## Data Compression

```python
import gzip

def compress_csv_file(input_file, output_file):
    with open(input_file, 'rb') as f_in, gzip.open(output_file, 'wb') as f_out:
        f_out.writelines(f_in)

# Example usage
input_file = 'daily_dataset_small_adjusted.csv'
output_file = 'daily_dataset_small_adjusted.csv.gz'
compress_csv_file(input_file, output_file)
```
