# Data Preprocessing

```python
import pandas as pd

# Read csv file
df = pd.read_csv('https://raw.githubusercontent.com/jabdurrahmaan/smart-meter-benchmark/main/datasets/daily_dataset_small.csv')

# Create column 'id' for primary key usage
df["id"] = df["LCLid"] + "_" + df["day"].astype(str)
columns = df.columns.tolist()
columns.insert(0, columns.pop(columns.index("id")))

df = df[columns]
df
```

```python
import json
data = df.to_dict(orient='records')

# Save json file locally
with open('daily_dataset_small_kv.json', 'w') as f:
    json.dump(data, f, indent=4)
```
